```python
import json

# Databricks egna LLM 
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

prompt = """
Generate a JSON array of 50 marathon results. Each object should have exactly these fields:
- "Year of birth": integer between 1960 and 2005
- "Gender": either "M" or "F"
- "Athlete country": 3-letter country code like "SWE", "USA", "KEN"
- "Event name": "Tokyo Ultra Marathon (2024) - 42km"
- "Event distance/length": "42km"
- "Athlete performance": time in format "H:MM:SS" for example "3:45:22"
- "Athlete average speed": float in km/h for example 11.2
- "Athlete ID": integer between 10000 and 99999

Return ONLY the JSON array, nothing else.
"""

response = w.serving_endpoints.query(
    name="databricks-meta-llama-3-1-70b-instruct",
    messages=[{"role": "user", "content": prompt}]
)

# Spara svaret som text
raw_text = response.choices[0].message.content
print(raw_text[:300])  # Skriv ut lite så ser att det funkar
```
