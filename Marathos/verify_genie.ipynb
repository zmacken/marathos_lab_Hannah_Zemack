Question: How many unique athletes are there?

Answer: There are 1,635,502 unique athletes in the database.

```sql
SELECT
  COUNT(athlete_id) as unique_athletes
FROM
  marathos.gold.dim_athlete
```

Question: What is the distribution of athletes by gender in mart_gender_distribution?

Answer: Gender Distribution of Athletes
The athlete population is heavily male-dominated, with men representing over 80% of all participants. 2
Key Statistics:
Male (M): 5,972,433 participants (80.86%)
Female (F): 1,413,913 participants (19.14%)
Other (X): 53 participants (<0.01%)
Unknown: 5 participants (<0.01%)

```sql
SELECT
  athlete_gender,
  SUM(antal_deltagare) as total_participants,
  ROUND(
    SUM(antal_deltagare)
      * 100.0
      / (
        SELECT
          SUM(antal_deltagare)
        FROM
          marathos.gold.mart_gender_distribution
      ),
    2
  ) as percentage
FROM
  marathos.gold.mart_gender_distribution
GROUP BY
  athlete_gender
ORDER BY
  total_participants DESC
```
